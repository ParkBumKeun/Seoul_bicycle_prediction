```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Activation

from keras.callbacks import EarlyStopping

from keras.callbacks import ModelCheckpoint

```

```python
data= pd.read_excel('count_binary_data.xlsx')
data=data.dropna()
del data['count2']
```

```python
np.random.seed(5)
count = data['count'].values
```

```python
# timestep 14

lookback = 14
lookback_num = lookback + 1

result = []
for index in range(len(count) - lookback_num):
    result.append(count[index: index + lookback_num])
```

```python

result= np.array(result)

# train 70%, test 70%
row = int(round(result.shape[0] * 0.7))
train = result[:row, :]

```

```python
# x 데이터프레임 array 변환

X_train = train[:, :-1]
X_train = np.reshape(X_train, (X_train.shape[0], 14, 1))
Y_train = train[:, -1]

X_test = result[row:, :-1]
X_test = np.reshape(X_test, (X_test.shape[0], 14,1))
Y_test = result[row:, -1]
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python


model = Sequential()

model.add(LSTM(32, return_sequences=True, input_shape=(14, 1), activation='relu'))
model.add(LSTM(32, return_sequences=False, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['acc'])


model.summary()
```

```python
callbacks_list= [EarlyStopping(monitor='val_loss',patience=100)] 
hist = model.fit(X_train, Y_train,
    callbacks=callbacks_list, 
    validation_data=(X_test, Y_test),
    batch_size=20,
    epochs=400)
```

```python
scores = model.evaluate(X_test,Y_test)
print("%s : %.2f%%"% (model.metrics_names[1], scores[1]*100)) # ###########################d왜 %% 두 개????
# epoch, lass ,acc 그래프
plt.figure(1,figsize=(10,5))
loss = hist.history['loss'] # loss에 값넣기.
val_loss = hist.history['val_loss'] # val_loss 값넣기.
acc = hist.history['acc'] # history에 정확도 acc 도 갖고 있음
val_acc=hist.history['val_acc']

epochs = range(1, len(loss)+1) # loss의 길이만큼  지정
plt.plot(epochs,loss, 'b', label='Training loss')
plt.plot(epochs,val_loss, 'g', label='Validation loss')


plt.title('Training/Validation loss and accuracy')
plt.xlabel("Epochs")
plt.ylabel('Loss')
plt.legend()
plt.show()

Y_predict = model.predict(X_test)
df= pd.DataFrame(Y_predict)
df.insert(1,'Y_test',Y_test)
df.to_csv('C:/Users/admin/Desktop/LSTM_predict.csv')
```

```python
plt.plot(epochs,acc, 'y--', label='Training accuracy')
plt.plot(epochs,val_acc, 'r--', label='Validation accuracy')
plt.title('Training/Validation loss and accuracy')
plt.xlabel("Epochs")
plt.ylabel('Loss')
plt.legend()
plt.show()
```
