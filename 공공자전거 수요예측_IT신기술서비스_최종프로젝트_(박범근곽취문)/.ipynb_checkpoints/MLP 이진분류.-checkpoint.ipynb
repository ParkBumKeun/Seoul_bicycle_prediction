```python
#0. 사용할 패키지 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

from keras.callbacks import EarlyStopping

from keras.callbacks import ModelCheckpoint

#랜덤시드 고정시키기   
np.random.seed(5)

#1. 데이터 준비
dataset= pd.read_excel('count_binary_data.xlsx')
dataset = dataset.values # 값으로 변환   

#2. 데이터셋 분할
X = dataset[1:, :6]
Y = dataset[1:, 6]

# 데이터 분할 함수
def division(data):
    train_size= int(len(data)*0.6) # train 6, val 2 , test 2로
    val_size = int(len(data)*0.8)
    data_train = data[0:train_size]
    data_val = data[train_size:val_size]
    data_test= data[val_size:]
    return data_train, data_val, data_test



X_train,X_val,X_test = division(X)
Y_train,Y_val,Y_test = division(Y)

#3. 모델 구성
model = Sequential()
model.add(Dense(6, input_dim=6, activation='relu')) 
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid')) 

#4.모델 학습과정 설정
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 

#5. 모델 학습 
callbacks_list= [EarlyStopping(monitor='val_loss',patience=10)] 
hist = model.fit(X_train, Y_train, epochs=500, batch_size=100,
                 callbacks=callbacks_list,
                 validation_data=(X_val,Y_val)) 

#6. 모델 평가
scores = model.evaluate(X_test,Y_test)
print("%s : %.2f%%"% (model.metrics_names[1], scores[1]*100)) 

#7. 예측한 값 저장
Y_predict = model.predict(X_test)
df= pd.DataFrame(Y_predict)
df.insert(1,'Y_test',Y_test)
df.to_csv('C:/Users/admin/Desktop/MLP_predict.csv')

# 모델 아키텍쳐 요약
model.summary() 


# epoch, lass ,acc 그래프
loss = hist.history['loss'] # loss에 값넣기.
val_loss = hist.history['val_loss'] # val_loss 값넣기.
acc = hist.history['acc'] 
val_acc=hist.history['val_acc']

epochs = range(1, len(loss)+1) # loss의 길이만큼  지정
plt.plot(epochs,loss, 'b', label='Training loss')
plt.plot(epochs,val_loss, 'g', label='Validation loss')
plt.plot(epochs,acc, 'y--', label='Training accuracy')
plt.plot(epochs,val_acc, 'r--', label='Validation accuracy')

plt.title('Training/Validation loss and accuracy')
plt.xlabel("Epochs")
plt.ylabel('Loss')
plt.legend()
plt.show()


```
